```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
import functools as ft
import csv 
import numpy as np
```

```python
file = open('GSAF5.csv')
df= pd.read_csv(file)
df.columns
```

```python
df
```

```python
#1.-Remove duplicates
#2.-Remove Missing Values
#3.-Remove Incorrect Values
#4.-Remove Nan values
#5.-Cleanind Text
#6.-Data Type Correction
#7.-Rename Columns
#8.-Create New DataFrames
#9.-Drop unnecessary columns
```

```python
#1.-
before = len(df)
df = df.drop_duplicates()
after = len(df)
print("Number of duplicates dropped: ",str(before-after))
```

```python
#2.- 
null_cols = df.isnull().sum()
null_cols[null_cols>0]
```

```python
drop_cols = list(null_cols[null_cols>1000].index)
df = df.drop(drop_cols, axis =1)
```

```python
null_cols2 = df.isnull().sum()
null_cols2[null_cols2>0]
```

```python
#3.- Sharks are not aggresive animals so all the Unprovoked attacks should've been provoked
df['Type'].value_counts().Unprovoked
```

```python
df['Type'] = df['Type'].replace(['Unprovoked'], "Provoked")
df
```

```python
#4
df = df.dropna()
df
```

```python
#5.-
df ["Name"] = df ["Name"].str.replace('male','-')
df ["Name"] = df ["Name"].str.replace('female','-')
df ["Year"] = df ["Year"].replace(0,'-')
df
```

```python
#6.- #All good.
df.dtypes
```

```python
#7.-
df = df.rename(columns ={'Case Number':"Case No.",
                        "Investigator or Source":"Source",
                        "pdf":"PDF",
                        "Case No.1":"Case No.1",
                         "Case No.2":"Case No.2"})
df
```

```python
#9.-Drop unnecessary columns
df = df.drop(columns = ["Case Number.1","Case Number.2"])
```

```python
df.to_csv('Clean Data')
```

```python
#8 df1 =  Number of atacks of top 1 country sorted by area f2 = Countrys with more Skark Attacks 
countries = []
for i in df['Country'].unique():
    countries.append(i)
countries
for i in df['Country']:
    countries.append(i)
```

```python
occur = df.groupby(['Country']).size().reset_index(name = "No. of Attacks")
df_countries = pd.DataFrame(occur)
df_countries = df_countries.sort_values(by=['No. of Attacks'],ascending=False)
df2 = df_countries.reset_index(drop=True )
df2.to_csv("Number of shark attacks per country")
df2
```

```python
n_df = df[df['Country']=="USA"]
occur2 = n_df.groupby(['Area']).size().reset_index(name = "No. of Attacks")
df_area = pd.DataFrame(occur2)
df_area = df_area.sort_values(by=['No. of Attacks'],ascending=False)
df1 = df_area.reset_index(drop=True )
df1.to_csv("Number of shark attacks per Area in USA")
df1
```
